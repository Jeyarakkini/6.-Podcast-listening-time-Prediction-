```python
from Univariate import Univariate
```

```python
quan
```

```python
descriptive=Univariate.descriptive(raw_df,quan)
descriptive
```

```python
lesser,greater=Univariate.findOutlier(descriptive,quan)
```

```python
lesser
```

```python
replaced_df=Univariate.replaceOutlier(raw_df,descriptive,lesser,greater,quan)
```
